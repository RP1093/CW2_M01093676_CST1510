###  Payroll Statement

Ask for: employee name, hours worked, hourly rate, federal tax rate, state tax rate.
Print a formatted payroll slip with gross pay, each deduction, and net pay (2 d.p.).

```python
# input from the user
print("=== Employee Payroll Statement ===")
name = input("Enter employee's name: ")
hours_per_week = int(input("Enter number of hours worked in a week: "))
hourly_pay = float(input("Enter hourly pay rate: "))
federal_tax = float(input("Enter federal tax withholding rate: "))
state_tax = float(input("Enter state tax withholding rate: "))
# Calculation (Gross Pay,Net_pay etc.)
gross_pay = hours_per_week*hourly_pay
fed_withholding = gross_pay*federal_tax
state_withholding = gross_pay*state_tax
total_deduction = fed_withholding+state_withholding
net_pay = gross_pay-total_deduction
# Output the results of the calculations
print("\n--- Payroll Slip ---")
print(f"Employee Name:", name)
print(f"\nHours Worked: {hours_per_week}")
print(f"Pay Rate: ${hourly_pay:.2f}")
print(f"Gross Pay: ${gross_pay}")
print(f"Deductions:",
      f"\tFederal Withholding({(federal_tax*100):.2f}%): ${fed_withholding:.2f}"
      f"\tState Withholding({(state_tax*100):.2f}%): ${state_withholding:.2f}",
      f"\tTotal Deduction: ${total_deduction:.2f}",
      sep="\n")
print(f"Net Pay: ${net_pay:.2f}")
print("------------------------")
```

#### Question 2:

The program prompts the user to enter a <u>three-digit number</u> and determines whether the user wins according to the following rules:
1. If the user input matches the lottery number in the exact order, the award is £10,000.
2. If all the digits in the user input match all the digits in the lottery number, the award is £3,000.
3. If one digit in the user input matches a digit in the lottery number, the award is £1,000.

```python
import random

lottery = str(random.randint(100, 999))  # Generate a random three-digit lottery number

user = input("Enter your three-digit lottery number: ")  # Ask the user to enter their guess

print("Lottery number is:", lottery)  # Show the lottery number

if user == lottery:
    print("Exact match! You win £10,000!")  # User's number exactly matches the lottery number
elif sorted(user) == sorted(lottery):
    print("All digits match! You win £3,000!")  # All digits match but in a different order
elif any(digit in lottery for digit in user):
    print("One digit matches! You win £1,000!")  # At least one digit matches
else:
    print("Sorry, no match. You win nothing.")  # No digits match
```

- I used the Python in-built random module to make a small lottery game. The randint() function is used to create a random three-digit number between 100 and 999. I changed this number to a string using str() so it can be compared easily with the user’s input. <br>

- I used the input() function to get a three-digit number from the user. Since input() gives the value as a string, both the user’s number and the lottery number are in the same data type, which makes comparing easier. <br>

- I used variables to store both the random lottery number and the user’s number so that they can be used later in the program. <br>

- I used if, elif, and else statements to check the different conditions and decide the prize:

- If the user’s number exactly matches the lottery number, they win £10,000.

- If all digits match but are in a different order, they win £3,000.

- If one digit matches, they win £1,000.

- If none of the digits match, the user wins nothing. <br>

- I used the sorted() function to check if all digits matched but in a different order, and the any() function to check if at least one digit matched. <br>

- Finally, I used the print() function to show the lottery number and the result to the user in a clear and simple way. <br>

#### Question 3

Write a function that returns the sum of all the elements in a specified row in a matrix using the following header: def sum_row(matrix, rowIndex): Write a test program that reads a 4-by-4 matrix and displays the sum of each row.




```python
def sum_row(matrix, row_index): # function to add up all numbers in one row
    return sum(matrix[row_index])  # return the total of that row
def main():
    matrix = [   # list with some numbers
        [10, 20, 30, 40],
        [5, 15, 25, 35],
        [2, 4, 6, 8],
        [9, 18, 27, 36]
    ]
    for i in range(len(matrix)):  # go through each row in the matrix
        print(f"Sum of row {i} = {sum_row(matrix, i)}") # show the sum for each row 
main()
```

I created a function called sum_row that takes a matrix and a row index as inputs. The function adds up all the numbers in the given row using the sum() function and returns the total. By creating this function, I can reuse the same code for any row in the matrix instead of writing the same logic repeatedly.

I used a multidimensional list to represent the matrix. This means the list contains other lists, where each inner list represents a row of numbers. This allows me to organize the data in a table-like structure, which makes it easier to access and perform calculations on specific rows.

In the main() function, I defined the matrix and then used a for loop to go through each row. For each row, I called the sum_row function and printed the result using f-strings. The f-strings allowed me to include both the row number and the sum in the output in a clear way.

By writing the sum_row function, I learned how functions can make code more organized and reusable. Instead of repeating the same code for each row, I can call the function whenever I need it. This reduces errors and makes the code cleaner.

#### Question 4

Password Validation with Multiple Criteria: Ask the user for a password and validate whether it:
- Has more than 8 characters.
- Contains at least one uppercase letter.
- Contains at least one lowercase letter.
- Contains at least one number.
- Contains at least one special character (@, #, $, etc.).
- Extra: If the password contains a space, print an error message

```python
import re

def validate_password(password):  # s check if password is valid
    if " " in password:  # check if there are spaces in the password
        print("Error: Password should not contain spaces.")  # tell user password can't have spaces
        return False  # stop here if password has space
    if len(password) <= 8:  # check if password is too short
        print("Error: Password must be more than 8 characters.")  # tell user it must be longer
        return False  # stop if length is not enough
    if not re.search(r"[A-Z]", password):  # check for at least one uppercase letter
        print("Error: Password must contain at least one uppercase letter.")  # inform user
        return False  # stop if no uppercase
    if not re.search(r"[a-z]", password):  # check for at least one lowercase letter
        print("Error: Password must contain at least one lowercase letter.")  # inform user
        return False  # stop if no lowercase
    if not re.search(r"\d", password):  # check for at least one number
        print("Error: Password must contain at least one number.")  # inform user
        return False  # stop if no number
    if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):  # check for special character
        print("Error: Password must contain at least one special character (@, #, $, etc.).")  # tell user
        return False  # stop if no special char
    print("Password is valid!")  # if all checks passed, password is okay
    return True  # return True if password is valid
user_password = input("Enter a password to validate: ")  # ask user to type their password
validate_password(user_password)  # call the function to check it
```

-I made use of the Python in-built input() function to get a password from the user. The input() function takes the user’s input as a string, which can then be stored in a variable for further use.

-I used variables to store the password entered by the user so that I could use it later to perform various checks. To help validate the password, I made use of the re module, which allows regular expressions to be used in Python. Regular expressions are very useful for searching and matching patterns within strings.

-In this program, I used if, elif, and else statements to check if the password meet different conditions. These included checking if the password had no spaces, was more than eight characters long, and contained at least one uppercase letter, one lowercase letter, one number, and one special character. The re.search() function was used to test each of these conditions by matching the required pattern in the string.

-To display the results, I used the print() function, which outputs the corresponding messages to the user. The program prints an error message when a password rule is broken and prints a success message when all rules are satisfied.

-By using regular expressions, I was able to make my code shorter and more efficient compared to checking each character one by one. The re module helped me handle complex text conditions with just a few lines of code, showing how powerful and flexible Python can be when modules are used effectively.

#### Question 5

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

```python
clothing_items = [
    "T-shirt","Jeans","Dress","Shoes","Hat","Socks","Jacket","Skirt","Blouse","Sweater",
    "Shorts","Scarf","Belt","Gloves","Coat","Boots","Sandals","Hoodie","Tank top","Cardigan",
    "Jeans","T-shirt","Dress","Boots","Blazer","Cap","Sneakers","Leggings","Shirt","Tie",
    "Polo","Socks","Jacket","Hat","Sweatpants","Dress","Sandals","Scarf","Belt","Blouse"] # list of clothing items 
with open("clothing_stock.txt", "w") as file: # open a text file 
    for item in clothing_items: # go through all the items 
        file.write(item + "\n") # write each item to the file 
print("clothing_stock.txt file created successfully!") # result that file has created 
```

```python
with open("clothing_stock.txt", "r") as file: 
    lines = [line.strip() for line in file if line.strip()] # remove extra space and empty line 
unique_items = set() # store all unique clothing items
duplicates = set() # store the duplicates found
# check for duplicates by looping
for item in lines:
    if item in unique_items:
        duplicates.add(item)  # add it to duplicates
    else:
        unique_items.add(item) # if not, add it to unique_items
if duplicates: # print the results
    print("Duplicate entries found:")
    for d in duplicates:
        print("-", d)
else:
    print("No duplicate entries found.")
    
```

- I used the Python open() function to open a file called clothing_stock.txt in read mode. Using the with statement makes sure the file is automatically closed after reading, which is a good practice and helps prevent errors.

- Before that, I created a list of clothing items, some of which were repeated, and wrote them into the file using open() in write mode. This made sure the file had data in it before reading it again.

- I read each line from the file and used the strip() function to remove extra spaces and blank lines. All the cleaned lines were saved in a list so that I could use them later to check for duplicate items.

- I created two sets — one called unique_items and one called duplicates. Sets only keep unique values, which makes it easier to find repeated items. I then used a for loop to go through each item in the list.

- If the item was already in unique_items, it meant the item was repeated, so I added it to duplicates. If the item was not in unique_items, I added it there to keep track of it.

- After checking all the items, the program looked at the duplicates set. If there were any duplicates, it printed “Duplicate entries found” and listed the repeated items. If there were no duplicates, it printed “No duplicate entries found.”

- This program helped me check for repeated stock items in a simple way. Using sets and a for loop made it easy to organize the data and find duplicates without having to check each item manually.
